```python
hours=[17,18,19,20,26,27,30,31,32,38,40,45,50,55]
fig, ax = plt.subplots(figsize=(4,4))

ax.hist(hours,bins=5,edgecolor='y');
```

```python
import numpy as np
import matplotlib.pyplot as plt
median_hour=np.median(hours)
bins=[10,20,30,40,50,60]
#fig, ax = plt.subplots(figsize=(6,6))
plt.hist(hours,bins=bins,edgecolor="black",color="blue",alpha=0.6)

plt.axvline(median_hour,color="black",ls='--',label="Median hour")
plt.legend();
```
